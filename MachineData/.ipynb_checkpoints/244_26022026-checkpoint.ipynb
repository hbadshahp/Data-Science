```python
import pandas as pd
import numpy as np
df = pd.read_csv("data_244_26022026.csv")


df_copy = df.copy()

df_copy['shutheight'] = (
    df_copy['shutheight']
        .replace(0, np.nan)
        .bfill()
)

df_final = (
    df_copy
        .groupby(['hour_ist', 'epc', 'shutheight'], as_index=False)
        .agg({'hourly_count': 'sum'})
)
df_final = df_final.sort_values('hour_ist')
df_final
```

```python
epc_value = "e00000000000000000000002"

df_filtered = df_final[df_final['epc'] == epc_value]

df_filtered
```

```python
df_final[df_final['epc'] == epc_value]['hourly_count'].sum()
```
