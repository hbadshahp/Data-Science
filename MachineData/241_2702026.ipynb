```python
import pandas as pd
df = pd.read_csv("data_241_27022026.csv")
```

```python
import numpy as np

df_copy = df.copy()

df_copy['shutheight'] = (
    df_copy['shutheight']
        .replace(0, np.nan)
        .bfill()
)

df_copy.head()
```

```python
df_final = (
    df_copy
        .groupby(['hour_ist', 'epc', 'shutheight'], as_index=False)
        .agg({'hourly_count': 'sum'})
)
```

```python
df_final.head()
```

```python
df_final = df_final.sort_values('hour_ist')
```
