```python
result = 10 + 2 * 3 # Multiplication happens first: 10 + (2 * 3) = 16
print(result)
```

```python
result = (10 + 2) * 3 # Parentheses first: (10 + 2) * 3 = 36
print(result)
```

```python
result = 2 ** 3 ** 2 # Right-to-left exponentiation: 2 ** (3 ** 2) = 2 ** 9
print(result)
```
