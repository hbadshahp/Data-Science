# Creating DataFrames

# From Python Lists

```python
import pandas as pd

data=[
    ["Alice",25],
    ["Bob",30],
    ["Charlie",35]
]

df = pd.DataFrame(data,columns=["Name","Age"])
print(df)
```

## From Dictionary of Lists

```python
data = {
    "Name":["Alice","Bob","Charlie"],
    "Age":[25,30,35]
}

df = pd.DataFrame(data)
print(df)
```

## From Numpy Arrays

```python
import numpy as np

arr = np.array([[1,2],[3,4]])
df = pd.DataFrame(arr,columns=["A","B"])
print(df)
```

## From CSV files

```python
df = pd.read_csv("data.csv")
print(df)
```

## From JSON

```python
df = pd.read_json("data.json")
print(df)
```
