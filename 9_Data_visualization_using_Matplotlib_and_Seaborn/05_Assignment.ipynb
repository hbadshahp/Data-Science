```python
# Try making a stackplot with your own weekly Schedule!
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
Work = [5,0,0,0,0,0,0]
Sleeping = [0,0,0,0,0,0,0]
Studying= [0,0,0,0,0,0,0]
Family_time= [0,0,0,0,0,0,0]
Housework= [0,0,0,0,0,0,0]

labels = ['Work', 'Sleeping', 'Studying', 'Family_time','Housework']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral','red']

plt.figure(figsize=(6,6))

plt.stackplot(days,Work,Sleeping,Studying,Family_time,Housework,labels=labels,colors=colors,alpha=0.8)
plt.legend(loc='upper left')
plt.title('Weekly Activity Tracker')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(True)
```
