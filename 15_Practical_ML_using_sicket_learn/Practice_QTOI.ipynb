```python
# import libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load YOur CSV file
data = pd.read_excel('iris-train.xlsx')
data.head()
```

```python
# Separate Features and Label
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
model = RandomForestClassifier()
model.fit(x,y)
```

```python
X = pd.read_excel('iris-test.xlsx').iloc[:,:-1]
predictions = model.predict(X)
print(predictions)
```
