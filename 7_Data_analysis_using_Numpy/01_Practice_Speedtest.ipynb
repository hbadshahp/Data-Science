```python
# using python list
import time

size = 1_000_000_0

l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x,y in zip(l1,l2)]
end = time.time()

print(add[0:10])

print(end - start)
```

```python
import numpy as np
import time

size = 1_000_000_0

l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1+l2
end = time.time()

print(add[0:10])
print(end - start)
```

## Compare the time taken to multiply two Python lists vs. two NumPy arrays

```python
# lists
import time

# python list
size = 1_000_000
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result = [x*y for x,y in zip(list1,list2)]
end = time.time()
print(f"Python list multiplication time: {end-start}")
```

```python
arr1 = np.array(list1)
arr2 = np.array(list2)

start = time.time()
result = arr1*arr2
end = time.time()
print(f"Numpy multiplication time: {end-start}")
```
