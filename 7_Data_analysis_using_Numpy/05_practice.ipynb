# Data Types in Numpy

## 1 Introduction to NYmpy Data typeS


```python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr.dtype)
```

## 2 Changing Data types

```python
arr = np.array([1.5,2.7,3.9])
print(arr.dtype)
```

```python
arr_int = arr.astype(np.int32)
print(arr_int)
print(arr_int.dtype)
```

# example: Downcasting to save memory

```python
arr_large = np.array([100000,200000,300000],dtype=np.int64)
arr_small = arr_large.astype(np.int32)
print(arr_small)
print(arr_small.dtype)
```

# 3 why data types matter in numpy

```python
arr_int64 = np.array([1,2,3],dtype=np.int64)
arr_int32 = np.array([1,2,3],dtype=np.int32)

print(arr_int64.nbytes)
print(arr_int32.nbytes)
```

## String Data Type in Numpy

```python
arr = np.array(['apple','banana','cherry'],dtype='U10')
print(arr)
```

## CompleX numbers

```python
arr = np.array([1+2j,3+4j,5+6j],dtype='complex128')
print(arr)
```

## Object Data type

```python
arr = np.array([{'a' : 1},[1,2,3],'hello'],dtype='object')
print(arr)
```
