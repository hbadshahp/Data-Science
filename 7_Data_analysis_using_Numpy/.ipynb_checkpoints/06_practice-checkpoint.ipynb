```python
import numpy as np
import time

arr = np.arange(1_000_000)
```

```python
start = time.perf_counter()

result_loop=[]
for num in arr:
    result_loop.append(num**2)

end = time.perf_counter()
print(f"Loop time {end-start}")
```
